```python
from IPython.display import Image
Image(filename = 'image//kucing.jpg', width = 224,height = 224)
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename,target_size = (224,224))
```

```python
plt.imshow(img)
```

```python
resize_img = image.img_to_array(img)
final_image = np.expand_dims(resize_img,axis =0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
resize_img.shape
```

##mobileV1

```python
prediction = mobile.predict(final_image)
```

```python
print(prediction)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results = imagenet_utils.decode_predictions(prediction)
```
